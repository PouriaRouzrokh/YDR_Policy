```python
import pandas as pd
import os
```

```python
df = pd.read_csv("/Users/pouria/Documents/Coding/YDR Policy Scraping/data/processed/scraped_policies_data.csv")
df.head()
```

```python
## A function remove unnecessary characters and extra spaces or newlines from a string
import re
def clean_string(string):
    return re.sub(r'[^a-zA-Z0-9\s]', '', string)
```

```python
SAVE_DIR = "/Users/pouria/Documents/Coding/YDR Policy Scraping/data/processed/scraped_policies"

df_new = df.copy()
df_new['policy_content_path'] = None
for index, row in df.iterrows():
    if row["contains_policy"] == True:
        policy_name = row['url'].split("/")[-1]
        policy_content = clean_string(row['policy_content'])
        policy_content_path = os.path.join(SAVE_DIR, policy_name + ".txt") 
        df_new.loc[index, 'policy_content_path'] = policy_content_path      
        with open(policy_content_path, "w") as f:
            f.write(policy_content)
    else:
        df_new.loc[index, 'policy_content_path'] = None
        print(f"Skipping {row['url']} because it does not contain a policy")


```

```python
df_new.to_csv("/Users/pouria/Documents/Coding/YDR Policy Scraping/data/processed/scraped_policies_data.csv", index=False)
```
